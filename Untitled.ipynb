**機械学習とは 

人が大量のデータを分析してルールを導き出し、モデルを構築する作業を行う代わりに、機械学習はデータから知識を引き出すためのより効率的な手段を提供する 
用途：メールスパムフィルタ、文字や音声認識、WEB検索エンジン、AlphaGoなど 

３種類の機械学習がある 
教師あり学習、強化学習、教師なし学習

教師あり学習・・・主な目標は、ラベル付けされたトレーニングデータからモデルを学習し、未知のデータや将来のデータを予測できるようにする 
　教師とは、望ましい出力信号（ラベル）がすでに判明しているサンプルの集合 
　教師あり学習では、「新しいラベルがどのカテゴリに分類されるか」を予測するモデルについてトレーニングできる 
　教師あり学習には、分類と回帰がある 
  
　分類は、過去の観測に基づき、新しインスタンスを対象としてクラスラベルを予測することが目的 
　　スパムメールフィルタ→スパムメールかスパムメールでないかの２分類 
  　文字認識→文字の画像を読み込んで、どの文字かを多クラス分類する 
　回帰は連続値を予測する 
　　ｙ＝ｆ（ｘ）みたいな感じか　勉強時間とテストの点数の相関関係を分析する際に使ったりする 

強化学習・・・主な目標は、環境とのやり取りに基づいて性能を改善するシステムを開発すること 
教師あり学習と似ているが、教師あり学習の場合はラベルが「正解」をあらわしていたが、強化学習の場合はラベルが「報酬」をあらわす 
AlphaGoとか 

教師なし学習・・・ラベル付けされていないデータや構造が不明なデータを扱う 
　結果変数や報酬関数がなくても、データの構造を調べることで意味のある情報を取り出すことができる 
 
クラスタリングによるグループの発見 
　クラスタリングは、大量の情報を意味のあるグループとして構造化できる探索的データ解析手法 
　同じクラスに分類されたオブジェクトは何らかの類似性を共有する一方で、多クラスに分類されたオブジェクトは相違が大きい 
 
マーケティングプログラムを開発する際に、顧客の関心によってクラス分類をすることで、顧客集団を発見できる 
 
教師なし学習では、次元削減という項目がある 
3次元のデータを2次元に圧縮する　x1,x2,x3というパラメータを持つデータを、z1,z2というパラメータに次元削減する 

**xNNについて 

〇〇 Neural Networkの意味 
 
DNN・・・Deep Neural Network 深層学習 
　今までの機械学習では、学習するための層が２～３であったが、この層をより多くしたもの 
　コンピュータの処理能力が向上したことにより実現可能となった 
  
CNN・・・Convolutional Neural Network 
　畳み込みニューラルネットワーク 
　主に画像認識で使われる 
　画像を3*3や4*4のブロックごとに読み込んでいく 
　事前処理として、読み込む画像サイズの周囲を処理するブロックの大きさ-1で0埋めする 
　CNNとDNNを組み合わせることで、Googleの画像認識等が実現している 
  
RNN・・・Reccurent Neural Network 
　再帰型ニューラルネットワーク 
　自然言語処理で主に使われる 
　まったく詳しくないので勉強したら補足します。
